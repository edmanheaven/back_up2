Prepare nth month column

```python
rm_month_df = merged_df2[['loan_account_no','loan_application_date']]
rm_month_df['repayment_start_month'] = merged_df2['repayment_start_date'].dt.to_period('M')
rm_month_df['report_month'] = merged_df2['report_date'].dt.to_period('M')
rm_month_df['bucket_MA'] = merged_df2['bucket']
rm_month_df['nth month'] = rm_month_df['report_month'] - rm_month_df['repayment_start_month']
rm_month_df['nth month'] = rm_month_df['nth month'].apply(lambda x: x.n)
rm_month_df['nth month'] = rm_month_df['nth month'] + 1
```

```python
rm_month_df['bucket_MA'] = rm_month_df.bucket_MA.fillna('missing')
```
